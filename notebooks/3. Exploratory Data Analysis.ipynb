# 1. Phân tích tổng quan

## 1.1. Ma trận tương quan

```python
numeric_cols = df1.select_dtypes(include=[np.number]).columns
df_numeric = df1[numeric_cols]

correlation_matrix = df_numeric.corr(method='spearman')
mask = np.zeros_like(correlation_matrix)
mask[np.triu_indices_from(mask)] = True
# Vẽ biểu đồ tương quan
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix,mask=mask, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix', fontsize=20, weight= 'bold')
plt.show()
```

## 1.2. Hướng trung tâm

```python
numerical_columns = df1._get_numeric_data()

num_plots = len(numerical_columns.columns)
fig, axes = plt.subplots(nrows=4, ncols=5, figsize=(16, 8))
axes = axes.flatten()

for i, col in enumerate(numerical_columns.columns):
    sns.boxplot(x=numerical_columns[col], ax=axes[i], palette='Set2')
    axes[i].set_title(f'Boxplot of {col}')
    axes[i].set_xlabel(None)

for i in range(num_plots, len(axes)):
    axes[i].remove()

plt.tight_layout()
plt.show()
```

```python
columns_to_plot = ['Price', 'Original_Price', 'Discount', 'Discount_Rate']
plt.figure(figsize=(16, 10))
for i, column in enumerate(columns_to_plot, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df1[column], kde=True)
    plt.title(f'Distribution of {column}',fontweight='bold',fontsize=16)
    plt.xlabel(column,fontweight='bold',fontsize=12)
    plt.ylabel('Frequency',fontweight='bold',fontsize=12)

plt.tight_layout(pad=2)
plt.show()
```

## 2.2. Những cửa hàng lớn có nhiều ưu đãi giảm giá không?


```python
store_pro=df.groupby("Store")['Name'].count()
store_dis=df.groupby("Store")['Discount_Rate'].mean()
store_pro_df = store_pro.reset_index()
store_dis_df = store_dis.reset_index()
result = pd.merge(store_pro_df, store_dis_df, on='Store')
result.columns = ['Store', 'Number_of_Products', 'Average_Discount_Rate']
result=result.sort_values(by='Number_of_Products',ascending=False).reset_index(drop=True).head(20)

bar_width = 0.4

index = np.arange(len(result))
plt.figure(figsize=(16, 8))

fig, ax1 = plt.subplots()

bars1 = ax1.barh(index, result['Number_of_Products'], bar_width, label='Number of Products', color='blue')

ax2 = ax1.twiny()
bars2 = ax2.barh(index + bar_width, result['Average_Discount_Rate'], bar_width, label='Average Discount Rate', color='green')

ax1.set_yticks(index + bar_width / 2)
ax1.set_yticklabels(result['Store'])
ax1.set_ylabel('Store')
ax1.set_xlabel('Number of Products', color='blue')
ax2.set_xlabel('Average Discount Rate', color='green')

for i, v in enumerate(result['Number_of_Products']):
    ax1.text(v+2, i, str(v), color='blue', va='center',fontsize=6,  fontweight='normal')

for i, v in enumerate(result['Average_Discount_Rate']):
    ax2.text(v+0.02, i + bar_width, f'{v:.2f}', color='green', va='center',fontsize=6, fontweight='normal')

plt.title('Top 10 của hàng có nhiều mặt hàng nhất')

plt.show()
```

- Tiki Trading đang bày bán nhiều mặt hàng nhất (~50%), kế đến là Nhà sách Fahasa (~25%)
- Các cửa hàng phổ biến có tỉ lệ giảm giá trung bình khá cao (phần lớn trên 20%)
- Hầu hết các sản phẩm hiện nay đều đang được bán với ưu đãi giảm giá đi kèm

## 2.8. Những yếu tố nào có ảnh hưởng đến tỉ lệ giảm giá?

```python
def find_correlated_features_numeric(dataframe, target):
    df_num_corr = dataframe.corr()[target]
    correlated_features = df_num_corr[abs(df_num_corr) > 0.1].sort_values(ascending=False)
    return list(correlated_features.index)

num_correlated_list = find_correlated_features_numeric(df_numeric, 'Discount_Rate')
feature_selected = num_correlated_list[1:]
feature_selected
```

```python
def show_regression(feature_selected, target, df):
  fig, ax = plt.subplots(round(len(feature_selected) / 4), 4, figsize = (20, 8))

  for i, ax in enumerate(fig.axes):
      if i < len(feature_selected):
          sns.regplot(x=feature_selected[i], y=target, data=df, ax=ax)

show_regression(feature_selected, 'Discount_Rate', df1[feature_selected + ['Discount_Rate']])
```

- Trong tất cả các biến tương quan với 'Discount Rate', Biến 'Discount' tương quan nhất và hầu hết các biến đều tương quan thuận.
